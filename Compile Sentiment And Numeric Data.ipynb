```python
import pandas as pd
import pandas as pd
from google.colab import drive
import os
import numpy as np
import pandas_ta as ta


drive.mount('/content/drive')
```

```python
nums = pd.read_csv('/content/drive/Shareddrives/Stat 453 Project/Data Sharing Preprocessing/scraped_data.csv')
nums.head()
```

```python
nums['Date'] = pd.to_datetime(nums['Date'])
nums = nums.set_index('Date')
nums.head()
```

```python
sent = pd.read_csv('/content/drive/Shareddrives/Stat 453 Project/Sentiment data/new_test.csv')
sent.head()
```

```python
sent['timestamp'] = pd.to_datetime(sent['timestamp'])
sent = sent.set_index('timestamp')
sent.head()
```

```python
def find_duplicate_index_values(df):
    """
    Finds and returns duplicate values in the DataFrame's index.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.Index: An Index containing the duplicate values.
    """
    duplicate_mask = df.index.duplicated(keep=False)
    duplicate_values = df.index[duplicate_mask].unique()
    return duplicate_values

# Example usage with the 'sent' DataFrame:
duplicate_dates_in_sent = find_duplicate_index_values(sent)

if not duplicate_dates_in_sent.empty:
    print(f"Duplicate dates found in 'sent' DataFrame index:\n{duplicate_dates_in_sent}")
else:
    print("No duplicate dates found in 'sent' DataFrame index.")
```

```python
if not duplicate_dates_in_sent.empty:
    # Get the first duplicate date from the list of duplicates
    first_duplicate_date = duplicate_dates_in_sent[0]
    # Filter the DataFrame to show all rows for that specific duplicate date
    duplicate_rows = sent.loc[first_duplicate_date]
    print(f"First two rows for the duplicate index '{first_duplicate_date}':")
    display(duplicate_rows.head(2))
else:
    print("No duplicate dates found in 'sent' DataFrame index.")
```

```python
# Find the full date range for the 'sent' DataFrame
min_date_sent = sent.index.min()
max_date_sent = sent.index.max()
full_date_range_sent = pd.date_range(start=min_date_sent, end=max_date_sent, freq='D')

# --- FIX START ---
# Remove duplicate indices before reindexing to avoid 'cannot reindex on an axis with duplicate labels' error
sent = sent[~sent.index.duplicated(keep='first')]
# --- FIX END ---

# Reindex the 'sent' DataFrame to include all dates in the range
sent_reindexed = sent.reindex(full_date_range_sent)

# Impute missing values with the previously populated value (forward fill)
sent_imputed = sent_reindexed.ffill()

print("Original 'sent' DataFrame head after reindexing (showing NaNs if any):")
display(sent_reindexed.head())
print("\nImputed 'sent' DataFrame head:")
display(sent_imputed.head())

print("\nOriginal 'sent' DataFrame tail after reindexing (showing NaNs if any):")
display(sent_reindexed.tail())
print("\nImputed 'sent' DataFrame tail:")
display(sent_imputed.tail())

# Verify that there are no more gaps in the imputed 'sent' DataFrame
date_index_sent_imputed = sent_imputed.index
differences_sent_imputed = date_index_sent_imputed.to_series().diff().dropna()
if (differences_sent_imputed == pd.Timedelta(days=1)).all():
    print("\nThe imputed 'sent' date index now has all consecutive days.")
else:
    print("\nThe imputed 'sent' date index still has gaps (this should not happen after ffill).")

sent = sent_imputed # Update the sent DataFrame with the imputed version
```

```python
merged_df = pd.merge(nums[['macro_oscillator']], sent, left_index=True, right_index=True, how='inner')
merged_df.index.name = 'Date' # Explicitly set the index name
display(merged_df.head())
display(merged_df.tail())
```

```python
# Calculate RSI
merged_df.ta.rsi(close='close', append=True, offset = 1)

# Calculate Bollinger Bands
merged_df.ta.bbands(close='close', append=True, offset=1)

# Calculate Moving Averages (e.g., SMA 10 and SMA 20)
merged_df.ta.sma(close='close', length=10, append=True, offset=1)
merged_df.ta.sma(close='close', length=20, append=True, offset=1)

merged_df.dropna(inplace=True)
merged_df.head()
merged_df.tail()
```

```python
merged_df.to_csv('/content/drive/Shareddrives/Stat 453 Project/3_head_data_final.csv')
```

```python
date_index = merged_df.index

# Calculate the difference between consecutive days
differences = date_index.to_series().diff().dropna()

# Check if all differences are exactly one day
if (differences == pd.Timedelta(days=1)).all():
    print("The date index has all consecutive days.")
else:
    print("The date index does NOT have all consecutive days.")

# Optional: display non-consecutive dates if any
non_consecutive_dates = date_index[1:][differences != pd.Timedelta(days=1)]
if not non_consecutive_dates.empty:
    print("\nNon-consecutive dates (start of gap):\n", non_consecutive_dates)
    # Calculate the end of the gap
    end_of_gap_dates = non_consecutive_dates + pd.Timedelta(days=1)
    print("\nEnd of gap dates:\n", end_of_gap_dates)
```
